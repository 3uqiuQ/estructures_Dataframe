```python
print(len(data['CarrierDelay']),len(data['WeatherDelay']),len(data['NASDelay']),len(data['SecurityDelay']),len(data['LateAircraftDelay']))  
```

```python
data['CarrierDelay'] = data['CarrierDelay'].fillna(0)  
data['WeatherDelay'] = data['WeatherDelay'].fillna(0)  
data['NASDelay'] = data['NASDelay'].fillna(0)  
data['SecurityDelay'] = data['SecurityDelay'].fillna(0)  
data['LateAircraftDelay'] = data['LateAircraftDelay'].fillna(0)  
Ret_Comp = sum(data['CarrierDelay'])  
Ret_Meteo = sum(data['WeatherDelay'])  
Ret_Aerop = sum(data['NASDelay'])  
Ret_Seg = sum(data['SecurityDelay'])
# (Este de momento no lo uso porque es el retraso del vuelo anterior acumulado -creo-)
Ret_Ret = sum(data['LateAircraftDelay'])
Ret_Total = Ret_Comp + Ret_Meteo + Ret_Aerop + Ret_Seg  
```

Como me interesa saber el retraso acumulado por cada causa, he sustituido los valores NaN por 0 y he hecho la suma de toda la columna, sabiendo así cuánto acumula cada retraso a lo largo del año.  
Se puede apreciar que practicamente todo el retraso es causa de las companyies o els aeroports.

```python
Ret_Comp_pc = round((Ret_Comp * 100) / Ret_Total, 2)
Ret_Meteo_pc = round((Ret_Meteo * 100) / Ret_Total, 2)
Ret_Aerop_pc = round((Ret_Aerop * 100) / Ret_Total, 2)
Ret_Seg_pc = round((Ret_Seg * 100) / Ret_Total, 2)

print('Retards per causa de la companyia: ' + str(Ret_Comp_pc) + ' %.')
print('Retards per causa del aeroport: ' + str(Ret_Aerop_pc) + ' %.')
print('Retards per causa de la meteorologia: ' + str(Ret_Meteo_pc) + ' %.')
print('Retards per causa de seguretat: ' + str(Ret_Seg_pc) + ' %.')
```

Hi ha diverses columnes on falta força informació rellevant, com el temps real transcorregut, de manera que en aquestes columnes, la informació no és fiable, per la qual cosa he optat per suprimir algunes. El dataset és força gran, així que no hauria de falsejar la informació general.

```python
SinNull = data.loc[(-data['ActualElapsedTime'].isnull()) & (-data['AirTime'].isnull())]
SinNull.isna().sum()
```

Treient les files on els valors de *ActualElapsedTime* i també de *AirTime* són nuls, el dataset queda net.

```python
SinNull = SinNull.copy()
SinNull['Gap'] = (SinNull['CRSElapsedTime']) - (SinNull['ActualElapsedTime'])
```

```python
SinNull['AvSpeed'] = round((SinNull['Distance']) / ((SinNull['AirTime']) / 60),2)
```

```python
Dist_Gap = 4000
SinNull['DistCode'] = ' '
ListDistCode = []

for i in SinNull['Distance']:
    if i <= Dist_Gap:
        ListDistCode.append('Short') 
    elif i > Dist_Gap:
        ListDistCode.append('Long')
    else:
        ListDistCode.append('NaN')
SinNull['DistCode'] = ListDistCode
```

```python
SinNull['Status'] = ' '
ListStatus = []

for i in SinNull['Gap']:
    if i <= 0:
        ListStatus.append('On Time') 
    else :
        ListStatus.append('Delayed')
SinNull['Status'] = ListStatus
```

He afegit algunes columnes que crec que poden ser interessants,
bé per ajudar a visualitzar el dataframe o bé per poder treballar-hi més endavant:
- La columna *Gap* És la diferència entre el temps real i l'estimat.
- La columna *DistCode* Divideix els vols entre Llarga i Curta diastància, on es veu que el volum de vols de llarga distància és molt petit en comparació amb els de curta, cosa que sembla que connecta amb el filtratge anterior on la majoria de retards es donaven als vols de curta distància.
- La columna *AvSpeed* És la velocitat mitjana de l'avió (A l'aire) i està expressada a Millas/Hora.
- La columna *Status* mostra si el vol s'ha endarrerit o no.
